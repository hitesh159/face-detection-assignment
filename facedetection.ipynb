```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('test2.jpg')
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(image_gray,cmap='gray')
```

```python
haar_cascade_face = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
haar_cascade_face
```

```python
#scalefactor is used to compensate for some faces being more prominent than others being closer to the camera
#minNeighbors specifies the number of neighbors a rectangle should have to be called a face

faces_rects = haar_cascade_face.detectMultiScale(image_gray, scaleFactor = 1.2, minNeighbors = 5)
```

```python
print('Faces found: ', len(faces_rects))
```

```python
#making rectangles around the found faces
for (x,y,w,h) in faces_rects:
     cv2.rectangle(image, (x, y), (x+w, y+h), (255, 0, 0), 2)
```

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```
